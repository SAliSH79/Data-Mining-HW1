```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.NaN, strategy='mean') # Imputing with mean value for Numerical Features
data.bill_length_mm = imputer.fit_transform(data[column_name[1]].values.reshape(-1,1))[:,0] #Reshaping data from array to tabular presentation
data.bill_depth_mm = imputer.fit_transform(data[column_name[2]].values.reshape(-1,1))[:,0]
data.flipper_length_mm = imputer.fit_transform(data[column_name[3]].values.reshape(-1,1))[:,0]
data.body_mass_g = imputer.fit_transform(data[column_name[4]].values.reshape(-1,1))[:,0]
data
```

```python
imputer = SimpleImputer(missing_values=np.NaN, strategy='most_frequent') #Imputing with most frequent values for Non Numerical features
data.island = imputer.fit_transform(data[column_name[0]].values.reshape(-1,1))[:,0]
data.sex = imputer.fit_transform(data[column_name[5]].values.reshape(-1,1))[:,0]
data.species = imputer.fit_transform(data[column_name[6]].values.reshape(-1,1))[:,0]
data
```
